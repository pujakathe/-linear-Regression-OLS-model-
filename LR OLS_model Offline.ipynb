### Separate the Numerical and Categorical variables
1. categorical variables:the variables which are having category should be converted to dummies variables
2. the linear regreassion will only accept number so we will make dummy variables

```python
df_num=df[["Age","Nodeg","Earnings_1974","Earnings_1975","Earnings_1978"]]
```

```python
df_num.head()
```

```python
df_dummies=pd.get_dummies(df[["Race","Hisp","MaritalStatus","Eduacation"]])
```

```python
df_dummies.head(5)
```

### Combine Numerical and dummy variables

```python
df_combine=pd.concat([df_num,df_dummies],axis=1)
```

```python
df_combine
```

## Splitting thr data  in  training and test set
 1. using skleran we split 70% of the data into trainig set and test abd rest in test set
 2. settind random_state will give the same trainig and set  everytime on runing the code

```python
x=df_combine.columns
x
```

```python
X=df_combine[['Age', 'Earnings_1974', 'Earnings_1975',
       'Race_NotBlack', 'Race_black', 'Hisp_NotHispanic', 'Hisp_hispanic',
       'MaritalStatus_Married', 'MaritalStatus_NotMarried',
       'Eduacation_HighSchool', 'Eduacation_Intermediate',
       'Eduacation_LessThanHighSchool', 'Eduacation_PostGraduate',
       'Eduacation_graduate']]
y=df_combine["Earnings_1978"]
```

```python
# random_state is the seed used by random number generator .it can be any integer
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=100)
```

```python
X_train
```

## Step 4:Performing Linear Regression

```python
 from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

##  Step 5: Model Evalution

```python
coeff_df=pd.DataFrame(lm.coef_,X.columns,columns=["Coefficient"])
```

```python
coeff_df
```

```python
print(lm.intercept_)
```

 1. from the above result we may infern that if Age  increases by 1 unit ,it will affect Earnings_1978 by -105.199 units

###  step 6: Predictions

```python
y_pred=lm.predict(X_test)
```

```python
y_pred
```

### Model Performance metrics

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
r_squared=r2_score(y_test,y_pred)
```

```python
from math import sqrt
rmse=sqrt(mse)
```

```python
print("mean_squared_error:",mse)
print("root_mean_squared_error:",rmse)
print("r_squard_value",r_squared)
```

#####  if the value of R2 is 0.463  then this suggested that 46.35% of the variaction  in Y can be explained with help of given explonatory variable in taht model .
##### in other words it explains the propotion of  variation in  the dependant  variable that is explained by the independent variables
###### on addition of new variable the error is sure to decrease,thus R square always increase whenever a new variable is added to our model .this may not describe the important of a variable
